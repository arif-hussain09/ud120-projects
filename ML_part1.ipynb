```python
def randomization(n):
    """
    Arg:
      n - an integer
    Returns:
      A - a randomly-generated nx1 Numpy array.
    """
    #Your code here
    return (np.random.random([n,1]))
    raise NotImplementedError
# randomization(10)
def operations(h, w):
    """
    Takes two inputs, h and w, and makes two Numpy arrays A and B of size
    h x w, and returns A, B, and s, the sum of A and B.

    Arg:
      h - an integer describing the height of A and B
      w - an integer describing the width of A and B
    Returns (in this order):
      A - a randomly-generated h x w Numpy array.
      B - a randomly-generated h x w Numpy array.
      s - the sum of A and B.
    """
    #Your code here
    A=np.random.random([h,w])
    B=np.random.random([h,w])
    s=A+B
    return A,B,s
    raise NotImplementedError

# a,b,s=operations(2,3)
# print(a)
# print(b)
# print(s)

def norm(A, B):
    """
    Takes two Numpy column arrays, A and B, and returns the L2 norm of their
    sum.

    Arg:
      A - a Numpy array
      B - a Numpy array
    Returns:
      s - the L2 norm of A+B.
    """
    #Your code here
    s=np.linalg.norm(A+B)
    return s
    raise NotImplementedError
# norm(np.random.rand(2,3),np.random.rand(2,3))
```

BASIC NEURAL NETWORK
z=tanh(w^Tx)

```python
def neural_network(inputs, weights):
    """
     Takes an input vector and runs it through a 1-layer neural network
     with a given weight matrix and returns the output.

     Arg:
       inputs - 2 x 1 NumPy array
       weights - 2 x 1 NumPy array
     Returns (in this order):
       out - a 1 x 1 NumPy array, representing the output of the neural network
    """
    # Your code here
    wT=np.transpose(weights)
    prod=np.matmul(wT,inputs)
    z=np.tanh(prod)
    out=z
    return out
    # return (np.tanh(np.dot(weights.T,inputs)))
    raise NotImplementedError
x=np.random.randint(1,10,(2,1))
w=np.random.randint(0,3,(2,1))
print("\n",x)
print("\n",w)
print(f"OUTPUT{neural_network(x,w)}")
```

```python
def scalar_function(x, y):
    """
    Returns the f(x,y) defined in the problem statement.
    """
    #Your code here
    if x<=y:
      return x*y
    else:
      return x/y
    raise NotImplementedError
# print(scalar_function(4,1))

def vector_function(x, y):
    """
    Returns the f(x,y) defined in the problem statement.
    """
    #Your code here
    return np.vectorize(scalar_function)(x,y)
    raise NotImplementedError

print(vector_function(np.random.randint(0,10,(2,2)),np.random.randint(0,10,(2,2))))
```

**CLASSIFICATION**


*   perceptron

```python
from sklearn.linear_model import Perceptron
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score


clf_1=Perceptron(max_iter=40,random_state=0)
clf_2=LinearSVC(random_state=0)

clf_1.fit(X_train,y_train)
clf_2.fit(X_train,y_train)

y_pred_1=clf_1.predict(X_test)
y_pred_2=clf_2.predict(X_test)

print("Perceptron: %.4f " % accuracy_score(y_test,y_pred_1))
print("LinearSVC: %.4f " % accuracy_score(y_test,y_pred_2))
```

**CLASSIFY DIGITS**

```python
from sklearn.datasets import load_digits

digits=load_digits()
X ,y=digits.data ,digits.target

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2 ,random_state=0)

fig ,ax =plt.subplots(3,3,figsize=(8,8))

for i in range(9):
  ax[i//3,i%3].matshow(digits.images[i],cmap="gray")

```

```python
X_train.shape
```

```python
clf=LinearSVC(C=1,random_state=0)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print(f" Acurray: {accuracy_score(y_test,y_pred)}")
```

```python
from sklearn.metrics import confusion_matrix

M=confusion_matrix(y_test,y_pred)
print(M)
fig ,ax=plt.subplots()
sns.heatmap(M,annot=True,fmt="d",ax=ax)
```

```python
from sklearn.model_selection import cross_val_score
clf=LinearSVC(C=1 ,random_state=0)
score=cross_val_score(clf,X_train,y_train,cv=5)

print(f"Mean: {np.mean(score)}\nStandard Deviation: {np.std(score)}")
```

```python
from sklearn.model_selection import GridSearchCV

clf = LinearSVC(random_state=0)
param_grid = {'C': 10. ** np.arange(-6, 4)}
grid_search = GridSearchCV(clf, param_grid=param_grid, cv=5, verbose=3, return_train_score=True)
```

```python
grid_search.fit(X_train,y_train)
```

```python
print(f"Best prams: {grid_search.best_params_}")
```

```python
print(f"Best Score: {grid_search.best_score_}")
y_pred=grid_search.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test,y_pred)}")
```
